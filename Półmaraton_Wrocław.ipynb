```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [[405, 291, 90, 26, 22, 30, 6], [43, 608, 536, 168, 50, 58, 7], [4, 133, 1016, 878, 316, 208, 10], [3, 17, 249, 1194, 1045, 971, 11], [0, 3, 23, 228, 617, 1388, 8], [0, 1, 17, 94, 228, 1940, 15]]
```

```python
better_time = [data[0][0], sum(data[1][0:2]), sum(data[2][0:3]), sum(data[3][0:4]), sum(data[4][5:7]), sum(data[5][0:6])]
all_runners = [sum(row) for row in data]
better_time, all_runners
```

```python
results = [
    [better_time[i], all_runners[i] - better_time[i], f"{(better_time[i]/all_runners[i] * 100):.2f}%"]
    for i in range(len(better_time))
]
```

```python
results_df = pd.DataFrame(results, columns=['Czas lepszy lub zakładany', 'Czas gorszy niż przewidywany lub DNF', 'Procent'])
results_df
```

```python
all = [sum(better_time), sum(all_runners)]
all_percent = sum(better_time)/sum(all_runners) * 100
all_percent = round(all_percent, 2)
print(f'Liczba biegaczy, którzy ukończyli w zakładanym czasie lub szybciej to {all_percent}%')
```

```python
columns=['Czerwony', 'Zielony', 'Pomarańczowy', 'Niebieski', 'Fioletowy', 'Żółty', 'DNF']
index=['Czerwony', 'Zielony', 'Pomarańczowy', 'Niebieski', 'Fioletowy', 'Żółty']
```

```python
df = pd.DataFrame(data, columns=columns, index=index)
```

```python
sns.heatmap(df, annot=True, fmt=".0f", cmap='mako_r')
plt.xticks(rotation=45)
plt.xlabel("Końcowa strefa czasowa")
plt.ylabel("Przewidywana strefa czasowa")
```

```python
data_perc = [sum(row) for row in data]
data_perc
```

```python
arr = np.array(data, dtype=float)
for i in range(len(arr)):
    arr[i] = arr[i] / data_perc[i]
df1 = pd.DataFrame(arr, columns=columns, index=index)
```

```python
sns.heatmap(df1, annot=True, fmt=".1%", cmap='mako_r')
plt.xticks(rotation=45)
plt.xlabel("Końcowa strefa czasowa")
plt.ylabel("Przewidywana strefa czasowa")
```
